```python
import math, random
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5000
h = 1 / 52
tm = 0.5
s0 = 30
sig = 0.3
r = 0.05
K = 30
v = np.empty(N)

```

```python
# np.random.seed(1234)
t_ = (np.array([range(52)]) + 1) * h

for i in range(N):
    eps = np.random.normal(0, 1, 52)
    eps_ = eps.cumsum()
    s_path = s0 * np.exp((r - 0.5 * sig ** 2) * t_ + sig * np.sqrt(h) * eps_)
    if not (i % 50):
        plt.plot(np.insert(s_path[0], 0, 30))
    v[i] = max(s_path.mean() - K, 0)

p_option = v.mean() * np.exp(-0.05 * 0.5)
print(p_option)

```
